# Decision Tree Regression - Ice Cream Sales




## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('/content/Ice_cream selling data.csv')
X = dataset.iloc[:, :1].values
y = dataset.iloc[:, -1].values
```

```
plt.scatter(X, y, color = 'red')
plt.title('Ice Cream Sales vs Temperature')
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.show()
```

## Training the Polynomial Regression model on the whole dataset

```
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X, y)
```

## Visualising the Polynomial Regression results

```
plt.scatter(X, y, color = 'red')
plt.plot(X, regressor.predict(X), color = 'blue')
plt.title('Ice Cream Sales vs Temperature')
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.show()
```

## Predicting a new result with Decision Tree Regression

```
# Reshape the input to be a 2D array-like with a single sample
input_value = np.array([[3.7]])

# Predict using the decision tree regressor
predicted_value = regressor.predict(input_value)

print(predicted_value)
```
